1. В чем принципиальные отличия item-item рекомендаций от ALS?  
ALS относится к коллаборативным фильтрациям, item-item - content-based детерминированным алгоритмам. ALS использует эмбеддинги данных о пользователе и товаре, игнорируя свойства самого товара. В рекомендациях item-item используется информация о товаре и подбирается матчинг фич пользователей и товаров  

2. Приведите 3 примера весов user-item матрицы для задачи рекоммендаций товаров  
Развесовку матрицы можно проводить на основе любых метрических данных, пересекающихся между пользователям и товаром, основным критерием подбора является обеспечение необходимых показателей метрик качества. Например, за вес принять отчет времени в секундах, с момента просмотра товара до момента покупки при факте её свершения, если это удовлетворит условиям бизнес-процесса. Как правило, весами выступают: оценки товаров, выставляемые пользователями (1), факт прикладного взимодействия с товаром (2), количество комментариев (3).  

3. Какие ограничения есть у ALS? (Тип информации, линейность/нелинейность факторов и т д)  
Ограничения ALS:  
    По типу информации на выходе оценка релевантности товара, теоретически, диапазон может быть любым, информация на выходе модели, как правило, не имеет прямой связи с метриками качества  
    модель является линейной относительно факторов  
    отсутствие персонализации накладывает органичение по выстраиванию модели с учетом особенностей групп пользователей  
    модель не учитывает влияние нестационарных составляющих временного ряда  
    модель требовательна к однородности и качеству входных данных (как и многие другие)  
    имеет одинаковые веса ошибок предсказания
